```python
df1=df.copy()
df1
```

#### Encoding the categorical feature

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
dummies = pd.get_dummies(df1[['thal']])
df1 = pd.concat([df1,dummies],axis=1)
df1
```

```python
df1.drop({'thal','patient_id'},axis=1,inplace=True)
```

```python
df1.head()
```

#### Splitting the data

```python
X=df1[['slope_of_peak_exercise_st_segment','resting_blood_pressure','chest_pain_type','num_major_vessels','fasting_blood_sugar_gt_120_mg_per_dl','resting_ekg_results','serum_cholesterol_mg_per_dl','oldpeak_eq_st_depression','sex','age','max_heart_rate_achieved','exercise_induced_angina','thal_fixed_defect','thal_normal','thal_reversible_defect']]
y=np.array(df1[['heart_disease_present']])
y=np.ravel(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.22, random_state=9)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler() #initialise standard scaler
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.fit_transform(X_test)
```

#### KNN algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(X_train_scaled,y_train)
```

```python
y_predict=model.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score,recall_score, precision_score,f1_score,confusion_matrix
```

```python
print('accuracy score:',accuracy_score(y_true=y_test,y_pred=y_predict))
print('recall score:',recall_score(y_test,y_predict))
print('precision score:',precision_score(y_test,y_predict))
print('f1 score:',f1_score(y_test,y_predict))
print('confusion matrix:',confusion_matrix(y_test,y_predict))
```

```python
pd.crosstab(y_test,y_predict)
```

```python
test_error_rates=[]
for k in range(1,10):
    model=KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train_scaled,y_train)
    y_predict=model.predict(X_test_scaled)
    
    test_error=1-accuracy_score(y_test,y_predict)
    test_error_rates.append(test_error)
    
test_error_rates 
```

```python
plt.figure(figsize=(10,5))

plt.plot(range(1,10), test_error_rates , marker  = '*')
plt.xlabel('k neighbors')
plt.ylabel('test error')
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model= LogisticRegression()
```

```python
log_model.fit(X_train_scaled,y_train)
```

```python
model_predictions=log_model.predict(X_test_scaled)
model_predictions
Prob=log_model.predict_proba(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,precision_score,recall_score,f1_score,classification_report
```

```python
confusion_matrix(y_true=y_test,y_pred=model_predictions)
```

```python
pd.crosstab(y_test,model_predictions)
```

```python
print('accuracy_score:',accuracy_score(y_true=y_test, y_pred=model_predictions))
print('precision_score:',precision_score(y_true=y_test,y_pred=model_predictions))
print('recall:',recall_score(y_true=y_test,y_pred=model_predictions))
print('f score:',f1_score(y_true=y_test,y_pred=model_predictions))

print('-'*45)
print(classification_report(y_true=y_test,y_pred=model_predictions))
```
